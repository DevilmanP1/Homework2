```python
class Chocolate:
    def __init__(self, price: float, addition: str = None):
        if price <= 0:
            raise ValueError('Цена должна быть больше нуля')
        self.price = price
        self.addition = addition

    def show_my_chocolate(self):
        if self.addition:
            print(f'Шоколадный батончик, добавка: {self.addition}')
        else:
            print('Обычный шоколадный батончик')

    def __str__(self):
        return f'Шоколадный батончик, {self.price} рублей'


# Демонстрация использования класса
if __name__ == '__main__':
    try:
        # Создаем несколько объектов
        choco1 = Chocolate(100, 'орехи')
        choco2 = Chocolate(50)
        choco3 = Chocolate(150, 'клубника')

        # Используем метод show_my_chocolate
        choco1.show_my_chocolate()  # С добавкой
        choco2.show_my_chocolate()  # Без добавки
        choco3.show_my_chocolate()  # С добавкой

        # Используем print (вызывает __str__)
        print(choco1)
        print(choco2)
        print(choco3)

        # Проверка исключения
        # choco4 = Chocolate(-10.0)  # Вызовет ValueError

    except ValueError as e:
        print(f'Ошибка: {e}')
```
